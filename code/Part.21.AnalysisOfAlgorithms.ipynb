# 第B章 算法分析

算法分析是计算机科学的一个分支，主要研究算法的性能，特别是其运行时间和所需空间。

算法分析的实际目标是预测不同算法的性能，以指导设计决策。

首席执行官问奥巴马，对一百万个32位整数排序的最高效的方法是什么。很显然，有人已经提前告知了奥巴马，因为他立刻回答道，我认为冒泡排序起码是最差的方法。

但如果你是在一个像这样的采访中遇到这个问题，我认为最好的答案是，“对一百万个整数进行排序的方法是，用我正在使用的编程语言所提供的任意排序算法先实现。其性能对于大部分应用来说已经足够了，但如果我的应用程序确实太慢，我会用性能分析器来查看时间耗费在哪里。如果更快的排序算法对性能确实影响显著，那我会用一个较好的基线排序算法去实现。”

文涛：这又是一次艺术的体现。别去管最好的方式是什么，是先把它做到。做到了它就已经对大部分情况够用了，如果程序实在太慢，就再进行分析和优化。我想到这个场景可以用在面试中，虽然这样的方式可能会被面试官误解，但这也算是一次过滤，如果他不认可这样的方式，或许他是固定型心态，也可能很少去学习和读编程书了。至少在张亮这里，我能学得很好，我能随心所欲的做我的事情。如果我能让东哥信服，能让顾总相信我的工作态度，我可以在这里工作好几年，用来培养和训练自己。
