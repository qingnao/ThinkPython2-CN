**13.2 随机数**

多数程序，给定同样的输入，每次都是同样的输出，这就叫做确定性。确定性是件好事，因为我们总是希望同样的计算产生同样的结果。但对于某些应用来说，我们需要计算机能变得不可预测。游戏就是一个很好的例子，也有很多其他的应用场景。

想让一个程序完全随机，是很难的，但是可以让其看起来随机。一种方法是使用算法生成伪随机数。伪随机数不是真正随机的，因为它们是用一个确定的运算生成的，但只是从数字来看，很难与真随机进行区分。

random 模块提供了生成伪随机数的函数。函数 ramdom 会返回0.0到1.0之间的一个随机浮点数。不包含1.0

```python
import random

# for i in range(10):
#      x = random.random()
#      print(x)
    
# 函数 randint 接收参数 low 和 high， 返回 low 和 high 之间的一个整数（包括两者）.
random.randint(5, 10)

# 若想从一个序列随机选择一个元素，可以用 choice 函数： 
# choice 选择，抉择;
t = [1, 2, 3]
random.choice(t)

# random 模块也提供了些可基于某种连续分布，比如高斯分布，指数分布，伽马分布等，生成随机数的函数。
```

```python
import random
import string

def process_file(filename, skip_header):
    """
    读取文件，制作包含文件中单词的直方图。
    
    filename:文件地址
    skip_header：是否跳过书籍开头部分的标题信息。
    
    returns:从每个单词映射到它出现的次数
    """
    hist = {}
    fp = open(filename)
    
    if skip_header:
        skip_gutenberg_header(fp)
        
    for line in fp:
        if line.startswith('*** END OF THIS'):
            # 当执行到结尾时退出执行
            break
        process_line(line, hist)
        
    return hist


def skip_gutenberg_header(fp):
    """
    这个函数的目的是跳过书籍文件的头部信息（即Gutenberg 的版权和附加信息），直到找到文件的起始位置。
    一旦找到这个行，break就会退出循环，不再继续检查接下来的行，文件读取将继续从这个标记之后开始。
    
    fp:打开文件对象
    """
    for line in fp:
        if line.startswith('*** START OF THIS'):
            break
    
    
def process_line(line, hist):
    """
    将一行中的单词添加到直方图中
    
    line：一行中的单词
    hist：直方图（从单词到频率的映射）
    """
    # 将分隔前的连字符 '-' 替换为空格 ' '
    line = line.replace('-', ' ')
    strippables = string.punctuation + string.whitespace
    
    for word in line.split():
        # 移除标点符号并将单词转换成小写
        word = word.strip(strippables)
        word = word.lower()
        
        # 修改直方图
        hist[word] = hist.get(word, 0) + 1
        
        
def most_common(hist):
    """
    按词频降序列出词频对
    
    hist: 从单词到频率的映射
    returns: 返回'（频率， 单词）对'  的列表
    """
    t = []
    for word, freq in hist.items():
        t.append((freq, word))
    
    t.sort()
    # 对列表 t 进行排序。默认情况下，Python会根据元组的第一个元素（即value，也就是词频）进行升序排序。（小的在上面）
    t.reverse()
    # 列表反转，使得排序后的列表从升序变成降序（即从最高词频到最低词频到顺序）
    return t


def subtract(d1, d2):
    """
    返回一个字典，其中包含 d1 中出现但 d2 不出现的所有键
    
    d1: 书中出现的所有单词
    d2: 单词表中的词汇
    """
    res = {}
    for word in d1:
        if word not in d2:
            # 如果d1中的单词不存在d2里，就写入字典中
            res[word] = None

    return res

def total_words(hist):
    """返回直方图中单词频率出现的总和"""
    return sum(hist.values())


def different_words(hist):
    """返回直方图中不同单词的数量"""
    # 这里只需要计算hist字典的大小就得到了不同单词的数量。
    # 而这这种封装更显得编程设计成熟。只用方法名就表达了它的作用，而且也方便后续扩展。
    return len(hist)


def random_word(hist):
    """
    从直方图中随机选择一个单词
    
    每个单词出现的概率与直方图中的频率成正比
    """
    t = []
    for word, freq in hist.items():
        # 这里会创建一个列表，长度是 freq， 出现freq个单词的集合。然后extend会将集合逐个添加进 t
        t.extend([word] * freq)
    
    return random.choice(t)
    
        
    
def main():
    hist = process_file('158-0.txt', skip_header=True)
    print('书中单词总数量:', total_words(hist))
    print('每个单词出现的次数:', different_words(hist))

    # 按词频降序列出词频对
    t = most_common(hist)
    print('书中出现次数最多的20个单词：')
    for freq, word in t[0:20]:
        # '\t' 是制表符，代表着四个空格，也就是一个tab
        print(word, '\t', freq)

    words = process_file('words.txt', skip_header=False)

    # 取出所有出现在书中，但不在单词表中的词汇。
    diff = subtract(hist, words)
    print('书中出现的单词但不在单词表中的词汇列表是：')
    for word in diff:
        print(word, end='')

    print('\n\n这些是书中出现的一些随机单词')
    for i in range(100): # 输出100个
        print(random_word(hist), end=' ')

main()
```

```python
 # 13.4 最常用单词
 # 若想找到最常用的单词，我们可以创建一个元组列表，每个元组都包含一个单词和其频数，然后据此将列表排序。
 
 # 下面函数接收一个词频统计，然后返回一个词-频元组列表：
def most_common(hist):
    t = []
    for key, value in hist.items():
        t.append((value, key))
    
    t.sort(reverse=True) # 直接倒序排序
    return t 


hist = process_file('158-0.txt', skip_header=True)

# 每个元组中，频数在前，所以列表便是按频数排序。下面循环会输出前十个最常用单词：
t = most_common(hist)
print('The most common words are:')
# t[:10] 列表切片，从第一个取到第10个结束
for freq, word in t[:10]:
    print(word, '\t', freq)

# 此处使用关键词参数 sep，来令 print 用换表符而不是空格作为“分隔符”，这样第二类就会对齐。
# 下面是对 Emma 小说的统计结果：
# The most common words are:
# to 	 5223
# the 	 5183
# and 	 4822
# of 	 4279
# a 	 3113
# i 	 2861
# her 	 2445
# it 	 2395
# was 	 2393
# she 	 2334


  
# 13.5 可选参数
# 我们已经见到了诸多具有可选参数的内置函数和方法。而且我们也可以编写具有可选参数的自定义函数。
# 比如，下面是输出词频统计中最常用单词的函数：

def most_common(hist):
    t = []
    for key, value in hist.items():
        t.append((value, key))
        
    t.sort(reverse=True) 
    return t 

def print_most_common(hist, num=10):
    t = most_common(hist)
    print('The most common words are:')
    for freq, word in t[:num]:
        print(word, freq, sep='\t')

# 第一个参数是不可或缺的；第二个是可选的。num的默认值是10.
# 如果你只给定一个参数：
print_most_common(hist)
# num会使用默认值。如果给定两个参数：
print_most_common(hist, 20)
# num会使用实参来赋值。换句话说，可选参数会覆盖默认值。
# 如果一个函数同时有必须和可选参数，则必须参数在前，可选参数在后。
```

```python
# 13.5 字典减法
# 从本书选择不在 words.txt 中的单词会比较麻烦，你可能已经意识到，这是个集合减法；
# 也就是说，我们希望找出所有在某集合（本书单词），但不再另一个集合（单词列表）中的单词。

# subtract 函数会接受字典 d1 和 d2，返回一个新字典，其包含所有 d1中包含，但 d2中不包含的键。
# 因为我们不关心键的值，所以就将其都设为None.
def subtract(d1, d2):
    res = dict()
    for key in d1:
        if key not in d2:
            res[key] = None
    return res
```

```python
# 13.7 随机单词
# 若要从词频统计中选择随机单词，最简单的方法是每个单词复制一定数量，复制的数量为其对应的频数，合并为列表，然后从中选择单词。

def random_word(h):
    t = []
    for word, freq in h.items():
        t.extend([word] * freq)
    
    return random.choice(t)

# 表达式 [word] * freq 会将 word 复制 freq 份。extend 方法和 append 作用相似，只是前者参数是个序列。
# 这个额算法有效，但不够有效率；每次选择随机单词，都需要重新构建和原书一样大的列表。
# 显然，更优的方案是只建立一次列表，然后多次选择，但列表依然很大。

# 一个更高效的方案是：
# 1. 使用 keys 生成本书单词的列表。
# 2. 用词频的各步累计之和，构建列表。列表中最后一项，便是书中单词总数，n。
# 3. 选择一个从1 到 n 的随机值。使用二分搜索，寻找随机值在累计值列表中所嵌入位置的索引。
# 4. 使用该索引值从单词列表中确定对应的单词。
```

**13.9 数据结构**

目前来说，我们考虑更多的是实现难易，但在选择数据结构时，还需要考虑一些其他因素。比如运行时间。总有一些时候，理论上可以预期某种数据结构就是比另外一种要快。比如，我提到过的 in运算符，字典比列表快，最起码元素数量很多的时候，一目了然。

但往往你无法提前知晓哪种实现要更快。一种方法是，对每种数据结构都进行实现，然后运行比较，看哪种更快。这个方法叫做基准分析。更实际一些的方法是选择一种最容易实现的数据结构，然后看看是否满足预期。如果满足，便没必要优化了。如果不满足，可以使用一些工具，像profile模块，以定位程序最耗时的部分。

文涛：这一段又一次印证了完成比完美更重要。一开始在无法知晓哪种方法更好的时候，选择一种最容易实现的方式，看是否满足预期。我在编码过程中可以进行取舍，一方面考虑扩展的问题，一方面也考虑怎么将功能实现最快。

另一个需要考虑的因素是存储空间。但对于多数应用来说，运行时间的考量要优先于存储空间。

**13.10 调试**

调试一个程序时，尤其是面对严重问题时，一定要做好以下五个步骤：

- **阅读（Reading）：** 检查代码，读给自己，看看所写是否符合所想。
- **运行（Running）：** 做些修改，运行个版本程序。通常，如果你在程序合适的位置，输出合适的内容，很容易定位问题，不过有些时候，你还是需要辅以脚手架代码。（还是那句话，用脚手架可能会耗费点时间，但会减少大量调试的时间）
- **反思（Ruminating）：** 花点时间多思考！是哪种错误：语法，运行时，还是语义错误？从异常信息，或者程序输出中能得到什么信息？哪种错误会引发你看到的问题？问题出现前，你最后一次改动了什么？
- **小黄鸭（Rubberducking）** 如果你想另一个人阐述问题，有时候在问完问题之前，便找到了答案。通常，你不需要找别人；你只需要对一个橡皮鸭喃喃自语。这便是众所周知的`小黄鸭调试法`的来源。
- **回撤（Retreating）：** 有时候，最好的办法是回退，取消最近的修改，直到程序正常并且你能理解。然后开始再次构建。

新手程序员有时候会在以上某种步骤中徘徊，而忘记了其他步骤。上面的每一步都有其不灵的时候。

比如，遇到拼写错误时，阅读代码往往有效，但如果是概念误用，那阅读代码便无济于事了。如果你未理解代码的功能，那即使你阅读100次，也永远不会发现错误，因为你理解的就有问题。

运行实例往往有用，尤其是一些小而简单的测试。如果你不思考，也不阅读，便进行试验，那便会陷入我称之为“随机游走编程”的误区中，这种过程是随机修改，直至程序正常。可想而知，随机游走编程必然会耗费大量时间。

你要多花时间思考。调试就像实验科学。对于问题所在，你至少要有一种假设。如果有两种或者多种可能性，想办法设计个测试，排除其中可能假设。

但如果错误过多，或者尝试修复的程序过于庞大复杂，那即使再好的调试方案，都会失效。有时候，最好的选择便是会退代码，简化程序，直到其正常切你能理解。

新手程序员往往不愿意回退代码，因为他们无法忍受删除一行代码（即使是错误代码）。可以在删减代码前，将程序复制一份，这样会让你感觉好一些。在你想还原的时候，便可以一次性复制回来。

定位严重的异常，需要阅读，运行，思考，有时候还要回退代码。如果在某个步骤卡壳了，试试其他步骤。

文涛：新手程序员无法忍受删除一行代码，就像刚开始写作的人不愿意删除自己写的句子一样，哪怕是写得很一般的句子也不舍得删除，总觉得是用了很多心血写出的内容。大胆的删除它吧，让文章变的精简，变得内容紧凑，这样无论是自己还是别人读起来都会有一种享受感，而不是冗余凑字数的小作文。
